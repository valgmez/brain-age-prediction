```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import scipy.stats as st
%matplotlib inline
```

```python
# ---- Schizo
schizo_train = pd.read_csv('schizo/results_train.csv')
schizo_y_train = schizo_train.y_train
schizo_yhat_train = schizo_train.yhat_train
schizo_corrected_age_train = schizo_train.corrected_age_train

schizo_test = pd.read_csv('schizo/results_test.csv')
schizo_y_test = schizo_test.y_test
schizo_yhat_test = schizo_test.yhat_test
schizo_corrected_age_test = schizo_test.corrected_age_test
```

```python
# ---- Schizo Brain age delta (better than MAE because not absolute)
print('Schizo')
print('Brain age delta before correction (healthy): %.4f +/- %.3f' % (round((schizo_yhat_train - schizo_y_train).mean(), 4), np.std(schizo_yhat_train - schizo_y_train)))
print('Brain age delta after correction (healthy):  %.4f +/- %.3f' % (round((schizo_corrected_age_train - schizo_y_train).mean(),4), np.std(schizo_corrected_age_train - schizo_y_train)))

print('Brain age delta before correction (clinical): %.4f +/- %.3f' % (round((schizo_yhat_test - schizo_y_test).mean(), 4), np.std(schizo_yhat_test - schizo_y_test)))
print('Brain age delta after correction (clinical):  %.4f +/- %.3f' % (round((schizo_corrected_age_test - schizo_y_test).mean(),4), np.std(schizo_corrected_age_test - schizo_y_test)))
```

```python
# ---- Schizo Correlations
print('Schizo')
print('Correlation between uncorrected predicted age and chronological age (healthy): ',
    round(np.corrcoef(schizo_y_train, schizo_yhat_train)[0, 1], 4))
print('Correlation between corrected predicted age and chronological age (healthy): ',
    round(np.corrcoef(schizo_y_train, schizo_corrected_age_train)[0, 1], 4))
print('Correlation between uncorrected predicted age and chronological age (patients): ',
    round(np.corrcoef(schizo_y_test, schizo_yhat_test)[0, 1], 4))
print('Correlation between corrected predicted age and chronological age (patients): ',
    round(np.corrcoef(schizo_y_test, schizo_corrected_age_test)[0, 1], 4))

print()
print('Correlation between uncorrected brain age delta and chronological age (healthy): ',
    round(np.corrcoef(schizo_y_train, (schizo_yhat_train - schizo_y_train))[0, 1], 4))
print('Correlation between corrected brain age delta and chronological age (healthy): ',
    round(np.corrcoef(schizo_y_train, (schizo_corrected_age_train - schizo_y_train))[0, 1], 4))
print('Correlation between uncorrected brain age delta and chronological age (patients): ',
    round(np.corrcoef(schizo_y_test, (schizo_yhat_test - schizo_y_test))[0, 1], 4))
print('Correlation between corrected brain age delta and chronological age (patients): ',
    round(np.corrcoef(schizo_y_test, (schizo_corrected_age_test - schizo_y_test))[0, 1], 4))
```
